```python
S=dar[(dar['–ö–∞—Ç–µ–≥']=='C') & (dar['–ú–∞—Ä—à—Ä—É—Ç']=='–°–ê–õ–ï–•–ê–†–î')] #—á–∞—Å—Ç—å —Ç–∞–±–ª–∏—Ü—ã —Å —Ä–µ–π—Å–∞–º–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ C –∏ –º–∞—Ä—à—Ä—É—Ç–æ–º –°–ê–õ–ï–•–ê–†–î
S
```

```python
S.sort_values(by='Baggage Weight', ascending=False) #—Ä–µ–∑—É–ª—å—Ç–∞—Ç –∏–∑ –ø—É–Ω–∫—Ç–∞ –≤—ã—à–µ —Å —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–æ–π –ø–æ —É–±—ã–≤–∞–Ω–∏—é Baggage Weight
```

```python
sort=dar.groupby(['–ù–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ', '–¢–∏–ø –í–°'])[['–ú–í–ú, —Ç–æ–Ω–Ω','–í–º–µ—Å—Ç–∏–º–æ—Å—Ç—å']].mean() #–≤—ã–≤–µ–ª–∞ —Å—Ä–µ–¥–Ω—é—é –ú–í–ú –∏ –í–º–µ—Å—Ç–∏–º–æ—Å—Ç—å —Å –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–æ–π –ø–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é –∏ —Ç–∏–ø—É –í–°
sort
```
