```python
S=dar[(dar['Катег']=='C') & (dar['Маршрут']=='САЛЕХАРД')] #часть таблицы с рейсами категории C и маршрутом САЛЕХАРД
S
```

```python
S.sort_values(by='Baggage Weight', ascending=False) #результат из пункта выше с сортировкой по убыванию Baggage Weight
```

```python
sort=dar.groupby(['Направление', 'Тип ВС'])[['МВМ, тонн','Вместимость']].mean() #вывела среднюю МВМ и Вместимость с группировкой по направлению и типу ВС
sort
```
