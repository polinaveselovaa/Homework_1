```python
class Chocolate:

  def __init__(self, price: float, filling: str, nuts: bool, number: int):
    self.price=price
    self.filling=filling
    self.nuts=nuts
    self.number=number

  def __str__(self):
    if self.price < 0:
     raise ValueError('ОШИБКА! Цена батончика должна быть больше нуля!')
    else:
     return f'Шоколадный батончик, {self.price} рублей'

  def show_my_chocolate(self):
    if self.filling:
      print(f'Шоколадный батончик, добавка: {self.filling}')
    else:
      print('Обычный шоколадный батончик')

  def count_nuts(self):
    if self.nuts:
     sum = self.number * '*'
     print(sum)
    else:
      None
```

```python
bat1=Chocolate(-2.1, 'изюм', True, 5)
print(bat1)
```
